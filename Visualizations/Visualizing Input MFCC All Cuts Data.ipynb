```python
%matplotlib inline

import matplotlib.pyplot as plt 
import numpy as np
```

### Helper to subplot images

```python
def draw_10(arr):
    channels = arr.shape[3]
    fig, axes = plt.subplots(channels, 10, figsize=(10, 4))
    for i in range(channels):
        print "Channel ", i
        for j in range(10):
            axes[i][j].axis('off')
            reshaped = arr[j,:,:,i]
            axes[i][j].imshow(reshaped)
```
